## Discussion
**Interpretation of Results:**

After viewing these two datasets and creating visual representations of the data within them, it seemed quite clear where these two sets of data overlapped in the right ways. The Humboldt current runs up the western coastline of South America, bringing cool, nutrient rich water from the Southern ocean as well as causing coastal upwelling up the entire continent to create one of the most biologically productive and diverse regions in the world. This current flows all the way to the equator where it then turns Westward, flowing right through the region that has the highest sea surface carbon dioxide concentration in our global ocean. And if those two factors aren't enough, the Galapagos Islands lie almost right at that intersection point. Home to reefs of both warm and cold water corals, these islands have lost 80% of their original population due to increased CO2 levels in our oceans. This is the perfect location to study a carbon dioxide converting reaction.

**Limitations and Future Work:**

The primarily limitation of a study at this location is that the small coral population remaining here has developed a resistance to rising CO2 concentrations, which may cause our carbon dioxide reduction to have a greater effect on the reefs here than it normally should. However at the same time, by reducing the carbon dioxide levels and helping the reefs get back to their usual health, we will also be creating multiple entire reefs of carbon dioxide resistant coral, which if transplanted to other reefs across the planet could help bring other reefs back to full health as well. Future work that should be done for this study is to determine what CO2 conversion method would be most effective in the Galapagos reef ecosystem, as well as determining how to keep side effects to a minimum. Then the next step would be engineering a method for this reaction to occur autonomously through computers.
